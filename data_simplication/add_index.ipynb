```python
import pandas as pd

path = r'./input/702评分.xlsx'
```

```python
data_702 = pd.read_excel(path)
data_702
```

```python
data_702 = data_702.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
```

```python
data_702
```

```python
name_list = data_702['File Name']
name_list
```

```python
a = name_list[0]
a.split('-')
```

```python
def complete(num):
    num_str = str(num)
    length = len(num_str)
    if length == 1:
        num_str = '00' + num_str
    elif length == 2:
        num_str = '0' + num_str
    return num_str
```

```python
index_sort = []
for i in range(len(name_list)):
    temp = name_list[i].split('-')
    num_str = complete(temp[0])
    temp_2 = temp[2].split('.')
    num_2 = temp_2[0]
    whole_num_str = num_str + '-' + complete(num_2)
    index_sort.append(whole_num_str)

index_sort
```

```python
data_702['index_sort'] = index_sort
data_702
```

```python
data_702.to_excel(r'./output/data_702.xlsx', index=False)
```
